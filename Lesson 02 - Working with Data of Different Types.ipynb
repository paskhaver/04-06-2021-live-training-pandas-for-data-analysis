## Filtering by One Condition

```python
nba.tail()
```

```python
nba["Team"] == "Los Angeles Lakers"
```

```python
is_laker = nba["Team"] == "Los Angeles Lakers"
nba[is_laker]
```

```python
is_point_guard = nba["Position"] == "PG"

nba[is_point_guard]
```

## Numeric Filtering

```python
is_high_earner = nba["Salary"] > 35000000
nba[is_high_earner]
```

```python
is_old = nba["Birthday"] < "1982-01-01"
nba[is_old]
```

## Filtering with Multiple Conditions (AND + OR)

```python
nba.head()
```

```python
is_center = nba["Position"] == "C"
```

```python
is_rocket = nba["Team"] == "Houston Rockets"
```

```python
nba[is_center & is_rocket]
```

```python
nba[is_center | is_rocket]
```

## The `isin` Method

```python
# nba["Team"] == "Boston Celtics"

target_teams = nba["Team"].isin(["Boston Celtics", "Golden State Warriors", "Charlotte Hornets"])
nba[~target_teams]
```

## The `between` Method

```python
nba[nba["Salary"].between(10000000, 20000000)].head()
```

```python
nba[nba["Birthday"].between("1980-01-01", "1989-12-31")].head()
```
