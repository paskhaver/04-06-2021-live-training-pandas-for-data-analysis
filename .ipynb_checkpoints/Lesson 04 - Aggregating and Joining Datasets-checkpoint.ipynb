## Import pandas Library

```python
import pandas as pd
```

## Import Dataset

```python
sales = pd.read_csv("sales_by_employee.csv", parse_dates = ["Date"])
```

## Create a Pivot Table

```python
sales.head()
```

```python
# The two lines below are equivalent
sales.pivot_table(index = "Date")
sales.pivot_table(index = "Date", aggfunc = "mean")
```

```python
sales.pivot_table(index = "Date", aggfunc = "count")
```

```python
sales.pivot_table(index = "Date", aggfunc = "sum")
```

```python
sales.pivot_table(index = "Date", aggfunc = "sum", values = "Revenue")
```

## Add Columns to Pivot Table

```python
sales.head(3)
```

```python
sales.pivot_table(
    index = "Date", columns = "Name", values = "Revenue", aggfunc = "sum"
)
```

```python
sales.pivot_table(
    index = "Date", columns = "Name", values = "Revenue", aggfunc = "sum", fill_value = 0
)
```

```python
sales.pivot_table(
    index = "Date", columns = "Name", values = "Revenue", aggfunc = "sum", fill_value = 0, margins = True
)
```

## Melting a Dataset

```python
# Wide dataset
# Grows in width

# Tall dataset
# Grows in height


# Name  Region  Sales
#  Wii   NA      41.49
#  Wii   EU      29.02

video_game_sales = pd.read_csv("video_game_sales.csv")
```

```python
video_game_sales.head()
```

```python
video_game_sales.melt(
    id_vars = "Name",
    value_vars = ["NA"]
).head()
```

```python
video_game_sales.melt(
    id_vars = "Name",
    value_vars = ["NA", "EU", "JP", "Other"],
    var_name = "Region",
    value_name = "Sales"
)
```

## Joining Datasets Together

```python
week1 = pd.read_csv("restaurant/week_1_sales.csv")
week1.tail(3)
```

```python
week2 = pd.read_csv("restaurant/week_2_sales.csv")
week2.tail(3)
```

## Concatenation

```python
pd.concat([week1, week2])
```

```python
pd.concat([week1, week2], keys = ["Week 1", "Week 2"])
```

## Inner Join

![Inner Join](images/InnerJoin.png)

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
# Customers who came in both weeks
week1.merge(week2, how = "inner", on = "Customer ID").head()
```

```python
week1[week1["Customer ID"] == 155]
```

```python
week2[week2["Customer ID"] == 155]
```

```python
# Customers who came in both weeeks and ordered the same food item
week1.merge(week2, how = "inner", on = ["Customer ID", "Food ID"]).head()
```

## Outer Join

![Outer Join](images/OuterJoin.png)

```python
week1.merge(week2, how = "outer", on = "Customer ID")
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", indicator = True)
```

## Left Join

![Left Join](images/LeftJoin.png)

```python
week1.head(2)
```

```python
customers = pd.read_csv("restaurant/customers.csv")
customers.head(2)
```

```python
week1.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID")
```
